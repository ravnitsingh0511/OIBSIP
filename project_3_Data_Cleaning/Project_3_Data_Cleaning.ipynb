```python
# Project 3: Data Cleaning

## Objective
The goal of this project is to clean a real-world dataset by handling missing values, removing duplicates, standardizing data formats, and treating outliers to prepare the data for analysis.

## Dataset
Airbnb listings dataset containing information such as price, room type, availability, and reviews.

## Data Cleaning Steps

### 1. Missing Value Handling
- Numerical columns were filled using the mean.
- Categorical columns were filled using the mode.

### 2. Duplicate Removal
- Duplicate rows were identified and removed to avoid double-counting.

### 3. Data Consistency
- Text columns were standardized by removing extra spaces and converting text to lowercase.
- Invalid price values (≤ 0) were removed.

### 4. Outlier Treatment
- Extreme price outliers above the 99th percentile were removed to prevent skewed analysis.

## Output
A cleaned dataset (`cleaned_airbnb_data.csv`) ready for further analysis or modeling.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df= df.drop_duplicates()
```

```python
df["price"].dtype
```

```python
df["price"].describe()
```

```python
df = df[df["price"]>0]
```

```python
upper_limit = df["price"].quantile(0.99)
df = df[df["price"] <= upper_limit]
```

```python
df["price"].describe()
```

```python
df.to_csv("cleaned_airbnb_data.csv", index=False)
```
